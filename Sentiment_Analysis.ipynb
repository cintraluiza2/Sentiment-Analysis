```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="cardiffnlp/twitter-roberta-base-sentiment-latest")
```

```python
from transformers import pipeline


def classificar_sentimento(input_usuario):
    """
    Função para classificar o sentimento com base no input fornecido pelo usuário.

    Parâmetros:
    - input_usuario (str): Texto que será analisado pelo modelo.

    Retorno:
    - dict: Contém a classificação (positiva, neutra, ou negativa) e a pontuação associada ao sentimento.
    """
    # Validando o input do usuário
    if not isinstance(input_usuario, str) or not input_usuario.strip():
        raise ValueError("O input deve ser uma string não vazia.")

    # Classificando o sentimento usando o pipeline
    resultado = pipe(input_usuario)

    # Exibindo o rótulo retornado pelo modelo para debug
    print(f"DEBUG - Resultado completo do modelo: {resultado}")

    # Retornando o resultado diretamente (modelo já fornece rótulos legíveis)
    return {
        "classificacao": resultado[0]["label"],  # Rótulo diretamente do modelo
        "confianca": resultado[0]["score"]
    }

# Exemplo de uso
if __name__ == "__main__":
    texto = "I don't have an opinion."
    sentimento = classificar_sentimento(texto)
    print(f"Classificação: {sentimento['classificacao']}, Confiança: {sentimento['confianca']:.2f}")
```
